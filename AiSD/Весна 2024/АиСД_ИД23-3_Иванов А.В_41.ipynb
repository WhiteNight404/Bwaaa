```python
# Задание 41 (а,б,в)
class Product:
    def __init__(self, name, price, quantity, expiration_date):
        self.name = name 
        self.price = price  
        self.quantity = quantity  
        self.expiration_date = expiration_date 

# Хеш-таблица для хранения объектов класса Продукт
products_table = {}

# Добавление продукта в хеш-таблицу
def add_product(product):
    products_table[product.name] = product  

# Нахождениу наиболее часто встречающегося ключа в хеш-таблице
def most_frequent_key(hash_table):
    frequency = {}  # Словарь для подсчета частоты встречаемости ключей
    max_count = 0  # Переменная для хранения максимальной частоты
    max_key = None  # Переменная для хранения ключа с максимальной частотой
    for key in hash_table:
        frequency[key] = frequency.get(key, 0) + 1  # Увеличиваем частоту ключа
        if frequency[key] > max_count:  # Если текущая частота больше максимальной
            max_count = frequency[key]  # Обновляем максимальную частоту
            max_key = key  # Обновляем ключ с максимальной частотой
    return max_key  

# Класс для хранения информации о заказах в ресторане
class Order:
    def __init__(self, order_number, dishes, total_cost, delivery_address):
        self.order_number = order_number  
        self.dishes = dishes  
        self.total_cost = total_cost  
        self.delivery_address = delivery_address 

# Класс хеш-таблицы с методом цепочек
class HashTable:
    def __init__(self, size):
        self.size = size  # Размер хеш-таблицы
        self.table = [[] for _ in range(size)]  # Инициализация пустых списков для цепочек
    def get_hash(self, key):
        return key % self.size  # Хеш-функция: остаток от деления ключа на размер таблицы
    def insert(self, key, value):
        hash_key = self.get_hash(key)  # Получаем хеш ключа
        key_exists = False  # Флаг для проверки существования ключа
        bucket = self.table[hash_key]  # Получаем цепочку по хешу ключа
        for i, kv in enumerate(bucket):
            k, v = kv
            if key == k:  # Если ключ уже существует в цепочке
                key_exists = True 
                break
        if key_exists:
            bucket[i] = ((key, value))  # Обновляем значение по ключу
        else:
            bucket.append((key, value))  # Добавляем новую пару ключ-значение
    def search(self, key):
        hash_key = self.get_hash(key)  # Получаем хеш ключа
        bucket = self.table[hash_key]  # Получаем цепочку по хешу ключа
        for k, v in bucket:
            if k == key:  # Если находим ключ
                # Форматируем вывод информации о заказе
                order_info = f"Заказ номер: {v.order_number}; Блюда: {', '.join(v.dishes)}; Общая стоимость: {v.total_cost}; Адрес доставки: {v.delivery_address}"
                return order_info.replace("'", "")  # Удаление апострофов и возвращение информации
        return None  # Если ключ не найден, возвращаем None

# Функция для вывода информации о всех продуктах в хеш-таблице
def print_all_products(hash_table):
    for product_name in hash_table:
        product = hash_table[product_name]
        print(f"Продукт: {product.name}; Цена: {product.price}; Количество: {product.quantity}; Срок годности: до {product.expiration_date}")

# Функция для вывода информации о всех заказах
def print_all_orders(hash_table):
    for i in range(hash_table.size):
        for k, v in hash_table.table[i]:
            print(hash_table.search(k))

# Создаем хеш-таблицу для заказов
orders_table = HashTable(10)

# Добавление заказов в хеш-таблицу
orders_table.insert(1, Order(1, ['Пицца', 'Сода'], 700, '123 Проспект Мира'))
orders_table.insert(2, Order(2, ['Суши', 'Чай'], 1500, '456 Улица Ленина'))

# Добавление продуктов
add_product(Product('Молоко', 60, 60, '2024-07-10'))
add_product(Product('Хлеб', 40, 20, '2024-06-15'))
add_product(Product('Сыр', 400, 5, '2024-06-16'))
add_product(Product('Колбаски', 100, 35, '2024-06-11'))

# Нахождение наиболее часто встречающегося ключа
frequent_key = most_frequent_key(products_table)
print(f"Наиболее часто встречающийся продукт: {frequent_key}")

print_all_products(products_table)
print_all_orders(orders_table)
```
