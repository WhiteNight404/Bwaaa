```python
# Задание 35

class Student:
    def __init__(self, name, surname, group_number, grades):
        self.name = name  
        self.surname = surname  
        self.group_number = group_number  
        self.grades = grades  
        self.prev = None  # Инициализация указателя на предыдущий элемент в списке как None
        self.next = None  # Инициализация указателя на следующий элемент в списке как None

class DoublyLinkedList:
    def __init__(self):
        self.head = None  # Инициализия начала списка как None
        self.tail = None  # Инициализия конца списка как None

    def append(self, student):
        if not self.head:  # Пуст ли список
            self.head = student  # Если список пуст, устанавливает студента как начальный и конечный элемент
            self.tail = student
        else:
            self.tail.next = student  # Устанавливает следующий элемент для текущего конца списка
            student.prev = self.tail  # Устанавливает предыдущий элемент для добавляемого студента
            self.tail = student  # Обновляет конец списка на добавляемого студента

    def display(self):
        elements = []  # Пустой список
        current = self.head  # Начинает с начала списка
        #Проверка и добавление информации о студентах
        while current:  
            elements.append({  
                'Имя': current.name,
                'Фамилия': current.surname,
                'Номер группы': current.group_number,
                'Предметы и оценки': current.grades
            })
            current = current.next  # Переходит к следующему элементу в списке
        return elements 

# Задание 36
    # Определение функции calculate_average, принимающей двусвязный список dll1
    def calculate_average(dll1):
        total_sum = 0  # Инициализация переменной для хранения суммы всех оценок
        count = 0  # Инициализация счетчика для подсчета количества оценок
        current = dll1.head  # Начало итерации с головы списка
        while current: 
            for grade in current.grades.values():  # Итерация по всем оценкам текущего студента
                total_sum += grade  # Добавление оценки к общей сумме
                count += 1 
            current = current.next 
        # Возвращение среднего арифметического, если есть оценки, иначе 0
        return total_sum / count if count != 0 else 0
    
# Задание 37
    def remove_duplicates(self):
        seen = set()
        current = self.head
        prev = None
        while current is not self.tail:  # Цикл для обхода элементов списка
            identifier = (current.name, current.surname, current.group_number)  # Идентификатор уникальности
            if identifier in seen:
                # Удаление текущего элемента
                prev.next = current.next
                current.next.prev = prev
            else:
                seen.add(identifier)
                prev = current
            current = current.next
            # Проверка, не вернулись ли мы к началу списка
            if current.next is self.head:
                break
        # Проверка последнего элемента списка
        if (self.tail.name, self.tail.surname) in seen:
            self.tail = prev
            self.tail.next = self.head
            self.head.prev = self.tail

# Пример использования
dll = DoublyLinkedList()
# # Здесь должен быть код для добавления студентов в список
# dll.remove_duplicates()

dll = DoublyLinkedList()  # Создает новый экземпляр двусвязного списка
dll.append(Student('Ярослав', 'Ярикович', 'ИД23-32', {'Математика': 5, 'Литература': 5, 'Физика': 4, 'История': 4})) 
dll.append(Student('Ярослав', 'Ярикович', 'ИД23-32', {'Математика': 10, 'Литература': 10, 'Физика': 10, 'История': 10})) 
dll.append(Student('Ярослав', 'Ярикович', 'ИД66-99', {'Математика': 5, 'Литература': 4, 'Физика': 5, 'История': 3})) 
dll.append(Student('Виктор', 'Сидоров', 'ИД23-32', {'Математика': 3, 'Литература': 4, 'Физика': 5, 'История': 6})) 
dll1 = DoublyLinkedList()
dll1.append(Student('Боб', 'Бобов', 'ИД32-23', {'Математика': 1, 'Литература': 2, 'Физика': 3, 'История': 4})) 
dll.remove_duplicates()

students_info = dll.display()  
students_info1 = dll1.display() 
for student in students_info: 
    print(student) 
for student in students_info1: 
    print(student) 
average = calculate_average(dll1)
print(f"Среднее арифметическое всех оценок: {average}")
```
